# Social_Netwok_Ads_Prediction

In this project we will be working with a fake advertising data set, indicating whether or not a particular internet user clicked on an Advertisement on a company website. We will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* #### 'Daily Time Spent on Site': consumer time on site in minutes
* #### 'Age': cutomer age in years
* #### 'Area Income': Avg. Income of geographical area of consumer
* #### 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* #### 'Ad Topic Line': Headline of the advertisement
* #### 'City': City of consumer
* #### 'Male': Whether or not consumer was male
* #### 'Country': Country of consumer
* #### 'Timestamp': Time at which consumer clicked on Ad or closed window
* #### 'Clicked on Ad': 0 or 1 indicated clicking on Ad

* ### Creating the dependent and dependent feature sets

```python
X = ads[['Daily Time Spent on Site', 'Age', 'Area Income',
             'Daily Internet Usage', 'Male']].values
y = ads['Clicked on Ad'].values
```

* ### Splitting the data into training and test set

```python
from sklearn.model_selection import train_test_split
```

```python
# Creating a training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

* ### Feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Creating a scaler
sc=StandardScaler()
```

```python
# Fitting the scaler to trhe training set and transforming the training set
X_train=sc.fit_transform(X_train)
```

```python
# Transforming the test set
X_test=sc.transform(X_test)
```

## Building and Training the Model

Now let's Biuld the model using the **Random Forest** class from the **Scikitlearn** library. 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=15,criterion='entropy',random_state=101)
print(model)
```

* ### Training the model

```python
model.fit(X_train,y_train)
```

* ### Generating predictions

```python
y_pred=model.predict(X_test)
```

* ### Evaluating the model

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
con_mat=confusion_matrix(y_test,y_pred)
performance=classification_report(y_test,y_pred)
print("Confusion matrix: \n",con_mat)
print()
print(performance)
```

* ### So we can coonclude that our model is performing preety well with 95% accuracy.
